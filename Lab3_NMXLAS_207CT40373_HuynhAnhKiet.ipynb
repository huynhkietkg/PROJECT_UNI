1.5 Dilation or Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# In kích thước hình ảnh
print(data.shape)

# Thực hiện phẫu thuật nhị phân (binary dilation) lần đầu
d1 = nd.binary_dilation(data)

# Hiện thị hình ảnh đã phẫu thuật
plt.imshow(d1, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()

# Thực hiện phẫu thuật nhị phân (binary dilation) với 3 lần lặp
d2 = nd.binary_dilation(data, iterations=3)

# Hiện thị hình ảnh đã phẫu thuật
plt.imshow(d2, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1.6 Coordinate Mapping

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# In kích thước hình ảnh
print(data.shape)

# Lấy chiều cao và chiều rộng của hình ảnh
V, H = data.shape

# Tạo chỉ số cho hình ảnh
M = np.indices((V, H))

# Thiết lập giá trị cho d
d = 5
q = 2 * d * np.random.rand(*M.shape) - d

# Chuyển đổi chỉ số thành kiểu integer
mp = M + q.astype(int)

# Áp dụng map_coordinates để lấy giá trị pixel từ hình ảnh
dl = nd.map_coordinates(data, mp)

# Hiện thị hình ảnh đã xử lý
plt.imshow(dl, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

1.7 Biến đổi chung

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pyplot as plt

# Định nghĩa hàm GeoFun
def GeoFun(outcoord):
    a = 10 * np.cos(outcoord[0] / 10.0) + outcoord[0]
    b = 10 * np.sin(outcoord[0] / 10.0) + outcoord[1]
    return a, b

# Đọc hình ảnh từ file (chế độ grayscale)
data = iio.imread('world_cup.jpg', mode='L')  # Sử dụng mode='L' để đọc hình ảnh grayscale

# Áp dụng geometric transformation
dl = nd.geometric_transform(data, GeoFun)

# Hiện thị hình ảnh đã xử lý
plt.imshow(dl, cmap='gray')
plt.axis('off')  # Tắt trục
plt.show()
```

2.Đổi màu hai đối tượng này.

```python
import cv2
import numpy as np

# Đọc ảnh
image = cv2.imread('exercise/colorful-ripe-tropical-fruits.jpg')
result = image.copy()

# Giả sử biết tọa độ quả đu đủ và dưa hấu
# (x, y, w, h) cho từng quả — bạn cần điều chỉnh theo ảnh thật

# Vị trí quả đu đủ (papaya)
x1, y1, w1, h1 = 100, 100, 80, 100  # Thay bằng tọa độ thực tế
papaya = result[y1:y1+h1, x1:x1+w1]

# Vị trí quả dưa hấu (watermelon)
x2, y2, w2, h2 = 250, 120, 90, 90  # Thay bằng tọa độ thực tế
watermelon = result[y2:y2+h2, x2:x2+w2]

# Đổi màu quả đu đủ: ví dụ đảo màu (inverted color)
inverted_papaya = cv2.bitwise_not(papaya)
result[y1:y1+h1, x1:x1+w1] = inverted_papaya

# Đổi màu quả dưa hấu: ví dụ chuyển sang màu xanh dương (blue tint)
blue_watermelon = watermelon.copy()
blue_watermelon[:, :, 1:] = 0  # Tắt kênh G và R, chỉ giữ kênh Blue
result[y2:y2+h2, x2:x2+w2] = blue_watermelon

# Hiển thị ảnh kết quả
cv2.imshow("Modified Fruits", result)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Lưu ảnh kết quả
cv2.imwrite('exercise/modified-fruits.jpg', result)
```

3.Xoay 2 đối tượng này 1 góc 45 độ và lưu vào máy.

```python
import cv2
import numpy as np

# Đọc ảnh
image = cv2.imread('exercise/quang_ninh.jpg')
result = image.copy()

# Giả sử biết tọa độ ngọn núi và con thuyền (x, y, w, h)
# Bạn cần chỉnh lại theo đúng ảnh thực tế

# Vị trí ngọn núi
x1, y1, w1, h1 = 100, 100, 150, 120
mountain = image[y1:y1+h1, x1:x1+w1]

# Vị trí con thuyền
x2, y2, w2, h2 = 300, 200, 100, 60
boat = image[y2:y2+h2, x2:x2+w2]

# Hàm xoay ảnh nhỏ quanh tâm
def rotate_image(img, angle):
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(img, matrix, (w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_REFLECT)
    return rotated

# Xoay 45 độ
rotated_mountain = rotate_image(mountain, 45)
rotated_boat = rotate_image(boat, 45)

# Chèn lại vào ảnh gốc (chỉ ghi đè lên vùng cũ — nên đảm bảo xoay không vượt khung)
result[y1:y1+h1, x1:x1+w1] = rotated_mountain
result[y2:y2+h2, x2:x2+w2] = rotated_boat

# Lưu và hiển thị
cv2.imshow("Rotated Objects", result)
cv2.imwrite("exercise/quang_ninh_rotated.jpg", result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
import imageio.v2 as iio
import numpy as np
import cv2

# Danh sách hình ảnh có sẵn
image_files = {
    '1': 'exercise/colorful-ripe-tropical-fruits.jpg',
    '2': 'exercise/ha-long-bay-in-vietnam.jpg',
    '3': 'exercise/pagoda.jpg'
}

def choose_image(choice):
    if choice in image_files:
        return iio.imread(image_files[choice])
    else:
        print("Lựa chọn hình ảnh không hợp lệ.")
        return None

def translate_image(img, tx=50, ty=50):
    rows, cols = img.shape[:2]
    M = np.float32([[1, 0, tx], [0, 1, ty]])
    return cv2.warpAffine(img, M, (cols, rows))

def rotate_image(img, angle=45):
    rows, cols = img.shape[:2]
    center = (cols // 2, rows // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1)
    return cv2.warpAffine(img, M, (cols, rows))

def scale_image(img, scale=1.5):
    return cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_LINEAR)

def shrink_image(img, scale=0.5):
    return cv2.resize(img, None, fx=scale, fy=scale, interpolation=cv2.INTER_AREA)

def coordinate_map(img):
    rows, cols = img.shape[:2]
    map_x = np.zeros((rows, cols), np.float32)
    map_y = np.zeros((rows, cols), np.float32)

    for i in range(rows):
        for j in range(cols):
            map_x[i, j] = cols - j - 1  # lật ngang
            map_y[i, j] = rows - i - 1  # lật dọc

    return cv2.remap(img, map_x, map_y, cv2.INTER_LINEAR)

def show_image(title, img):
    if img.ndim == 3 and img.shape[2] == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Chuyển BGR -> RGB
    else:
        plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(title)
    plt.show()

def main():
    print("MENU:")
    print("T - Tịnh tiến")
    print("X - Xoay")
    print("P - Phóng to")
    print("H - Thu nhỏ")
    print("C - Coordinate Map")
    try:
        action = input("Chọn thao tác (T/X/P/H/C): ").strip().upper()
        print("Chọn hình ảnh:")
        print("1. Fruits")
        print("2. Ha Long Bay")
        print("3. Pagoda")
        choice = input("Nhập số 1-3: ").strip()

        image = choose_image(choice)
        if image is not None:
            if action == 'T':
                result = translate_image(image)
                show_image("Tịnh Tiến", result)
            elif action == 'X':
                result = rotate_image(image)
                show_image("Xoay", result)
            elif action == 'P':
                result = scale_image(image)
                show_image("Phóng To", result)
            elif action == 'H':
                result = shrink_image(image)
                show_image("Thu Nhỏ", result)
            elif action == 'C':
                result = coordinate_map(image)
                show_image("Coordinate Map", result)
            else:
                print("Lựa chọn thao tác không hợp lệ.")
        else:
            print("Không tải được hình ảnh.")
    except KeyboardInterrupt:
        print("\nQuá trình bị hủy bởi người dùng.")
    except Exception as e:
        print(f"Đã xảy ra lỗi: {e}")

if __name__ == "__main__":
    main()
```
